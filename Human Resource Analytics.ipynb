## M.Categorical Data Distribution Insights

### 1.Countplot

```python
categorical_variables = ["OverTime", "MaritalStatus", "JobRole", "Gender", "Dept", "BusinessTravel"]
num_plots = len(categorical_variables)
num_cols = 3 
num_rows = (num_plots - 1) // num_cols + 1
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))
fig.subplots_adjust(hspace=0.5)  
for i, column in enumerate(categorical_variables):
    row, col = i // num_cols, i % num_cols
    ax = axes[row, col]
    sns.countplot(data=do, x=column, ax=ax)
    ax.set_title(f"Distribution of {column}")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
for i in range(num_plots, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])
plt.show()
```

**Define the Categorical Variables to Visualize:**

- `categorical_variables` is a list containing the names of categorical variables you want to visualize.

**Calculate the Number of Rows and Columns for the Subplot Grid:**

- `num_plots` is calculated as the length of `categorical_variables`, representing the total number of categorical variables to be visualized.
- `num_cols` is set to 3, which specifies the number of columns in the subplot grid. You can change this value to control the layout.

**Create a Grid of Subplots:**

- `fig, axes` are used to create a grid of subplots with the specified number of rows and columns.
- `fig.subplots_adjust(hspace=0.5)` adjusts the vertical spacing between subplots for better readability.

**Iterate Through Categorical Variables and Plot Countplots:**

- A for loop iterates through the categorical variables.
- For each variable, the code calculates the row and col position in the subplot grid based on the current iteration index.
- An axis (ax) is selected for the current subplot.
- A countplot is created for the current categorical variable using `sns.countplot`. The `data=do` parameter specifies the DataFrame to use, and `x=column` specifies the current categorical variable to visualize.
- The title of the subplot is set to "Distribution of {column}", where {column} is the name of the categorical variable.
- The x-axis labels are rotated 45 degrees for better readability.

**Remove Empty Subplots If Needed:**

- After plotting all categorical variables, the code checks if there are empty subplots and removes them from the grid.

**Finally, `plt.show()` is used to display the grid of subplots with countplots for each categorical variable.**

The result is a neatly aligned set of countplots, each showing the distribution of a specific categorical variable. The number of columns in the grid can be adjusted as needed to accommodate the variables 

### 2.Piechart

```python
columns_for_pie = ['AttritionStatus', 'BusinessTravel', 'Gender', 'MaritalStatus', 'JobRole', 'OverTime']
num_plots = len(columns_for_pie)
num_cols = 3 
num_rows = (num_plots - 1) // num_cols + 1
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))
for i, column in enumerate(columns_for_pie):
    row, col = i // num_cols, i % num_cols
    ax = axes[row, col]
    pie_data = do[column].value_counts()
    ax.pie(pie_data, labels=pie_data.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Pie Chart of {column}')
    ax.axis('equal')
for i in range(num_plots, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])
plt.show()
```

### 3.Gender-Based Attrition Analysis

```python
plt.figure(figsize=(5, 5))
gend = do.groupby(['AttritionStatus', 'Gender']).size().unstack()
ax = gend.plot(kind='bar', stacked=False, colormap="cividis", edgecolor='black')
plt.title('Attrition Level Based on Gender')
plt.xlabel('AttritionStatus')
plt.ylabel('Count')
plt.xticks(rotation=0)
for gender in ax.patches:
    width, height = gender.get_width(), gender.get_height()
    x, y = gender.get_xy()
    ax.annotate(f'{height}', (x + width/2, y + height), ha='right', va='bottom')
plt.show()
```

This code generates a bar chart to visualize the attrition levels based on gender in the dataset. The dataset is grouped by "AttritionStatus" (whether an employee has experienced attrition or not) and "Gender" columns. The resulting counts are then plotted as a bar chart.

- The `figsize` parameter sets the size of the figure (the chart's dimensions).
- The `gender_att` variable stores the data after grouping by "AttritionStatus" and "Gender" and unstacking it for visualization.
- The `kind='bar'` argument in `gender_att.plot()` specifies that a bar chart should be created.
- The `colormap="cividis"` argument sets the color map for the bars.
- `edgecolor='black'` adds black borders to the bars for clarity.
- The title and labels for the X and Y axes are added for better understanding.
- Annotations on top of each bar display the count of employees in each category.

This chart helps provide insights into how attrition is distributed across different genders in the dataset. The "AttritionStatus" column indicates whether an employee has experienced attrition (e.g., "Yes" for attrition and "No" for no attrition), and the chart breaks down this information by gender.

### 4.Education-Based Attrition Analysis

```python
plt.figure(figsize=(8, 5))
edu = do.groupby(['AttritionStatus', 'Education']).size().unstack()
ax = edu.plot(kind='bar', stacked=False, colormap="cividis", subplots=False, edgecolor='black')
plt.title('Attrition Level vs Education Level')
plt.xlabel('AttritionStatus')
plt.ylabel('Count')
plt.xticks(rotation=0)
for education in ax.patches:
    width, height = education.get_width(), education.get_height()
    x, y = education.get_xy()
    ax.annotate(f'{height}', (x + width/2, y + height), ha='left', va='bottom')
plt.show()
```

This bar chart visualizes the relationship between employees' education levels and their attrition status. Each bar represents a different education level, and the chart shows how many employees fall into each attrition category (e.g., "Yes" or "No"). The edgecolor parameter is set to "black" to provide clear borders around the bars for better visual separation and readability. This visualization helps in understanding how attrition relates to the education level of employees, with the bars indicating the count of employees in each category.

The `alpha=0.5` argument makes the bars slightly transparent for better visualization, and `label` provides labels for the two datasets.

- The `plt.title("Distribution of Total Working Years")` line sets the title for the chart to "Distribution of Total Working Years."

This chart provides insights into the distribution of total working years among employees who have experienced attrition and those who haven't. It helps in understanding the typical working years of employees within these two groups and how they compare.

### 7. Department-Based Attrition Analysis

```python
dept_counts = do[do['AttritionStatus'] == 'Yes']['Dept'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Attrition Distribution by Department')
plt.show()
```

1. `dept_counts = do[do['AttritionStatus'] == 'Yes']['Dept'].value_counts()` filters the DataFrame `do` to include only rows where the 'AttritionStatus' is 'Yes' and then counts the occurrences of each unique department ('Dept') in these filtered rows. It calculates the number of employees who have attrited in each department.
2. `plt.figure(figsize=(8, 5))` sets the figure size to 8 units in width and 5 units in height for the upcoming pie chart.
3. `plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', startangle=140)` creates a pie chart. The `dept_counts` values are used as the data to be visualized in the chart. The `labels` parameter specifies the labels for each pie slice, which correspond to the department names. `autopct='%1.1f%%'` displays the percentage on each slice with one decimal place. `startangle=140` determines the starting angle of the pie chart.
4. `plt.title('Attrition Distribution by Department')` sets the title for the pie chart, indicating that it's showing the distribution of attrition by department.
This code generates a pie chart that visually represents the distribution of attrition within different departments. Each slice of the pie corresponds to a department, and the size of each slice represents the proportion of employees who have attrited from that department. The percentage of attrition in each department is displayed on its respective slice.
